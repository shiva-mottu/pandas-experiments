```python
cols = list(data.columns.values)
data = data[ cols[0:4] + [ cols[-1] ] + cols[4:12] ]

data.head(5)
```

```python
## Filtering data
new_data = data.loc[(data['quality'] == 6) & (data['fixed acidity'] == 7.0 ) & ( data['chlorides'] > 0.069)]

#new_data.to_csv("modified.csv")

new_data.reset_index(drop=True , inplace=True)

new_data
```

```python
##Conditional Changes
data.loc[ data['fixed acidity'] == 7.0 , 'alcohol'] = True
data.loc[ data['total']>115, ['chlorides','quality'] ] = ['Test 1' , 'Test 2' ]
```

```python
##mean

data.groupby(['quality']).mean().sort_values('residual sugar', ascending= False)
```
